* Linear Regression testing NBA data 

```r
nba_train_data = read.csv('NBA_train.csv')
nba_test_data = read.csv('NBA_test.csv')
```

```r
str(nba_train_data)
str(nba_test_data)
```

```r
# using differnce in point between player and opponent to calcualte the winnings

nba_train_data$points_diff = nba_train_data$PTS - nba_train_data$oppPTS

scatter.smooth(nba_train_data$points_diff, nba_train_data$W)

# calculate linear regression between the wins and point diff


win_regression_model = lm(W ~ points_diff, data=nba_train_data)
summary(win_regression_model)
```
