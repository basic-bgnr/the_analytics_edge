__Objectives__:
1. To find out which food has the highest level of sodium content
1. To find out which foods have sodium level higher that a specified value


```r
#create a new data frame from usda data
food_sodium_dataframe = data.frame(usda$Description, usda$Sodium)

#finds out the dataframe label corresponding to the maximum Sodium level
food_sodium_dataframe[which.max(food_sodium_dataframe$usda.Sodium),]

#lists the names of the row in the dataframe
names(food_sodium_dataframe)


#specifies the value of the sodium above which the foods needs to be listed.
sodium_limiting_value = 10000


#creates a subset of dataframe which satisfies the condition of Sodium>sodium_limiting_value
subset(food_sodium_dataframe, food_sodium_dataframe$usda.Sodium>sodium_limiting_value)


sodium_level_of = 'CAVIAR'
index_of = match(sodium_level_of, usda$Description)
food_sodium_dataframe[index_of, ]

#scatter plot between totalFat and Protein 
scatter.smooth(x=usda$TotalFat, y=usda$Protein, xlab='Fat', ylab='Protein', main='Protein vs Fat')

#histogram plot of Total Fat , the y axis is always the frequency of firstvarialble)
#xlim in the maximum and minimum range in the x-axis and breaks defines how fine grained the box should be(no of box between the xlim.
hist(usda$TotalFat,xlab='Fat', ylab='Freq.', main='Freq. vs Fat', xlim=c(0, 100))
hist(usda$TotalFat,xlab='Fat', ylab='Freq.', main='Freq. vs Fat', xlim=c(0, 100), breaks=2000)
```

__Contents__: 
1. `dataframe$attribute=value` : adds the attribute to the existing dataframe

1. use of `tapply` function: takes 3 argument and performs as- group Arg1 according to Arg2 (usually fixed repeating value) and applys arg3 on the grouped value


__Objectives__:
1. Adding variables to data frames.
1. Understanding relation between various dataframe attibutes

```r
#creates a limiting value to filter required value from the set value
#here the set value is the mean of the data
limiting_value_protein = mean(usda$Protein, na.rm = TRUE)
limiting_value_fats    = mean(usda$Protein, na.rm = TRUE)


#the following operations results in list of logical values 
filtered_value_protein = usda$Protein > limiting_value_protein
filtered_value_fats    = usda$TotalFat > limiting_value_fats


#the following primitive converts the `logical` value into `Integer` value
filtered_value_protein_numeric = as.numeric(filtered_value_protein)
filtered_value_fats_numeric    = as.numeric(filtered_value_fats)

#additional attributes can be added to the existing dataframes by using $ sign and then the attribute name before the assignment operator
usda$HighProtein = filtered_value_protein_numeric
usda$HighFats    = filtered_value_fats_numeric


str(usda)

#groups data into table with column name 0 and 1
table(usda$HighProtein)
table(usda$HighFats)


#cross tabulates data with four possible value (first argument is x-axis and second argument is y-axis in a matrix)
table(filtered_value_protein, filtered_value_fats)

tapply(usda$Protein, usda$HighFats, mean)
```
