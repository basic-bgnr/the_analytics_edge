````r
countries = c("nepal", "india", "america", "brazil", "china");
life_expectancy = c(74, 76, 65, 83, 79);

CountryData = data.frame(countries, life_expectancy);
CountryData$population = c(100, 134, 2432, 234234, 123);

countries = c("Australia", "Greece");
life_expectancy = c(23, 234);
population = c(234234, 234234);

newCountryData = data.frame(countries, life_expectancy, population);
all_countries = rbind(CountryData, newCountryData);

all_countries
````

````r

who_europe2 = subset(who, ChildMortality<5 & Region=="Europe", select=c(ChildMortality,Region, Country, Population, LifeExpectancy))
who_europe2
str(who_europe2)

write.csv(who_europe2, "specialized_europe.csv")
````

````r
summary(who_europe2)
````

````r
summary(who$ChildMortality)
#finds the index of the country with the least child mortality rate
which.min(who$ChildMortality)

#prints luxembourg as
who$Country[100]

#prints details about the entire row of data of Luxembourg
subset(who, Country=="Luxembourg")


#plots the relation between mortality and lifeexpectancy
plot(who$ChildMortality, who$LifeExpectancy)
````

````r
#calculating answers to the online questions 

summary(who$Over60)

#finds out the country with the minimum over60 population
country_name = who$Country[which.min(who$Over60)]

#prints details about the aforementioned country
subset(who, Country==country_name)


#finds out the country with the largest literacy rate
country_name1 = who$Country[which.max(who$LiteracyRate)]

#prints details about the aforementioned country
subset(who, Country==country_name1)



````

### Using the tapply and table function 

````r
avgMortality = tapply(who$ChildMortality, who$Region, mean)
avgMortality[which.min(avgMortality)]
````
